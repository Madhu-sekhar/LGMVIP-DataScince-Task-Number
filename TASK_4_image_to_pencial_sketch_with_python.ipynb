```python
image_rgb = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
image_rgb
```

```python
#converting into the rgb fortamte
image_rgb = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
cv2_imshow(image_rgb)
```

```python
inverted_image=255-image_rgb
cv2_imshow(inverted_image)
```

```python
#code for blurring the RGB_image
blurred_image = cv2.GaussianBlur(inverted_image, (15, 15), 0)
cv2_imshow(blurred_image)
```

```python
inverted_blurr=255-blurred_image
pencil_sketch = cv2.divide(image_rgb, inverted_blurr, scale=256.0)
cv2_imshow(pencil_sketch)
```
